```python
def read_butt(f):
    new_df = pd.read_csv(f, parse_dates =['Date/Time'], skiprows = 19, header = 0)
    return new_df
```

```python
def round_datetime(date_list):
    for i in range(len(date_list)):
        date_list[i] = date_list[i].replace(minute=0, second=0, microsecond=0)
    return date_list
```

```python
def parse_butt_filename(f):
    site = f[0]
    if site.lower() != 'h' and site.lower() != 'j':
        site = 'ERROR'
        print f
    soil_type = f[1]
    if soil_type == 'I':
        soil_type = '1'
    if soil_type != '1' and soil_type != '2':
        soil_type = 'ERROR'
    if 'M1' in f:
        monitor_station = 'M1'
    elif 'M2' in f:
        monitor_station = 'M2'
    else:
        monitor_station = 'ERROR'
        print f
    
    if 'rh' in f.lower():
        data = 'RH'
    elif 'temp' in f.lower():
        data = 'Temp'
    else:
        data = 'ERROR'
        print f

    if 'hm' in f.lower() or 'h_' in f.lower():
        treatment = 'H'
        habitat = 'C'
    elif 'hs' in f.lower():
        treatment = 'H'
        habitat = 'S'
    elif 'p' in f.lower():
        treatment = 'C'
        habitat = 'P'        
    else:
        treatment = 'ERROR'
        habitat = 'ERROR'
        print f
    return site, soil_type, monitor_station, treatment, habitat
```

```python
def collapse_by_day(df):
    new_df = pd.DataFrame()
    
    new_df[['RH','Temp']] = df.groupby(pd.TimeGrouper('1D'))[['RH','Temp']].mean()
#     print new_df
#     new_df['soil_m_max'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].max()
#     new_df['soil_m_min'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].min()
#     new_df['soil_m_std'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].std()
    return new_df
```
