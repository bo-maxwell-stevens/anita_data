```python
import pandas as pd
import numpy as np
import os
from os.path import isfile, join
import xlrd
```

```python
mypath = 'sensor_data'
deca_files = []
for f in os.listdir(mypath):
    if 'Decagon' in f and not f.startswith('_'):
        if 'Jornada' in f or 'Hill' in f:
            deca_files.append(join(mypath,f))
```

```python
deca_files
```

```python
def parse_deca_filename(f):
    site = f[0]
    soil_type = f[1]
    #an I in one of the filenames (soil type), converted to 1
    if soil_type == 'I':
        soil_type = 1
    monitor_station = f[2:4]
    replicate = f[5]
    return site, soil_type, monitor_station, replicate
```

```python
def parse_deca(f):
    new_df = pd.read_csv(f, skiprows = 3, header = None)
    return new_df
```

```python
def collapse_by_day(df):
    new_df = pd.DataFrame()
    new_df['soil_m_mean'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].mean()
    new_df['soil_m_max'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].max()
    new_df['soil_m_min'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].min()
    new_df['soil_m_std'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].std()
    return new_df
```

```python
final_df = pd.DataFrame()
df = pd.DataFrame()
for x in range(len(deca_files)):
    for f in os.listdir(deca_files[x]):
        if 'csv' in f and not f.startswith('.'):
            tmp_df = parse_deca(join(deca_files[x],f))
            tmp_df.columns = ['Date_time','1','2','3','4','5']
            tmp_df.to_csv('tmp.csv')
            tmp_df = pd.read_csv('tmp.csv', index_col = 0, parse_dates = ['Date_time'])
            tmp_df = tmp_df.set_index('Date_time')
            
            df = pd.DataFrame(index = tmp_df.index.values)
            
            df1 = df.copy()
            df1['soil_m'] = tmp_df['1']
            df1 = df1[df1['soil_m'].notnull()]
            day_df1 = collapse_by_day(df1)
            day_df1['Date_time'] = day_df1.index.values
            day_df1['treatement'] = 'Cr'
            day_df1['habitat'] = 'C'
            day_df1['site'], day_df1['soil_type'], day_df1['monitor_station'], day_df1['replicate'] = parse_deca_filename(f)
            
            df2 = df.copy()
            df2['soil_m'] = tmp_df['2']
            df2 = df2[df2['soil_m'].notnull()]
            day_df2 = collapse_by_day(df2)
            day_df2['Date_time'] = day_df2.index.values
            day_df2['treatement'] = 'C'
            day_df2['habitat'] = 'P'
            day_df2['site'], day_df2['soil_type'], day_df2['monitor_station'], day_df2['replicate'] = parse_deca_filename(f)
            
            df3 = df.copy()
            df3['soil_m'] = tmp_df['3']
            df3 = df3[df3['soil_m'].notnull()]
            day_df3 = collapse_by_day(df3)
            day_df3['Date_time'] = day_df3.index.values
            day_df3['treatement'] = 'H'
            day_df3['habitat'] = 'C'
            day_df3['site'], day_df3['soil_type'], day_df3['monitor_station'], day_df3['replicate'] = parse_deca_filename(f)
            
            df4 = df.copy()
            df4['soil_m'] = tmp_df['4']
            df4 = df4[df4['soil_m'].notnull()]
            day_df4 = collapse_by_day(df4)
            day_df4['Date_time'] = day_df4.index.values
            day_df4['treatement'] = 'H'
            day_df4['habitat'] = 'S'
            day_df4['site'], day_df4['soil_type'], day_df4['monitor_station'], day_df4['replicate'] = parse_deca_filename(f)
    
            final_df = final_df.append(day_df1, ignore_index = True)
            final_df = final_df.append(day_df2, ignore_index = True)
            final_df = final_df.append(day_df3, ignore_index = True)
            final_df = final_df.append(day_df4, ignore_index = True)
            print join(deca_files[x],f)
final_df.to_csv('by_day_Decagon.csv')
```
