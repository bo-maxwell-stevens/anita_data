```python
mypath = 'sensor_data'
deca_files = []
for f in listdir(mypath):
    if 'Decagon' in f and not f.startswith('_'):
        if 'Jornada' in f or 'Hill' in f:
            deca_files.append(join(mypath,f))
```

```python
def parse_deca_filename(f):
    site = f[0]
    soil_type = f[1]
    #an I in one of the filenames (soil type), converted to 1
    if soil_type == 'I':
        soil_type = 1
    monitor_station = f[2:4]
    replicate = f[5]
    return site, soil_type, monitor_station, replicate
```

```python
def parse_deca(f):
    new_df = pd.read_csv(f, skiprows = 3, skip_blank_lines=True, header = None)
    return new_df
```

```python
final_df = pd.DataFrame()
df = pd.DataFrame() #columns = ['Date_time','site','soil_type','monitor_station','replicate','treatment','habitat','soil_m']
for x in range(len(deca_files)):
    for f in listdir(deca_files[x]):
        if 'csv' in f:
            tmp_df = parse_deca(join(deca_files[x],f))
            df['Date_time'] = tmp_df[0]
            df['site'], df['soil_type'], df['monitor_station'], df['replicate'] = parse_deca_filename(f)
            df1 = df
            df1['treatement'] = 'Cr'
            df1['habitat'] = 'C'
            df2 = df
            df2['treatement'] = 'C'
            df2['habitat'] = 'P'
            df3 = df
            df3['treatement'] = 'H'
            df3['habitat'] = 'C'
            df4 = df
            df4['treatement'] = 'H'
            df4['habitat'] = 'S'
            final_df = final_df.append(df1[df1['Date_time'].notnull()], ignore_index = True)
            final_df = final_df.append(df2[df2['Date_time'].notnull()], ignore_index = True)
            final_df = final_df.append(df3[df3['Date_time'].notnull()], ignore_index = True)
            final_df = final_df.append(df4[df4['Date_time'].notnull()], ignore_index = True)
            print join(deca_files[x],f)
```
