```python
def parse_deca_filename(f):
    site = f[0]
    soil_type = f[1]
    #an I in one of the filenames (soil type), converted to 1
    if soil_type == 'I':
        soil_type = 1
    monitor_station = f[2:4]
    replicate = f[5]
    return site, soil_type, monitor_station, replicate
```

```python
def read_deca(f):
    new_df = pd.read_csv(f, skiprows = 3, header = None, index_col = 0)
    return new_df
```

```python
def parse_date(xdate):
    finalIndex = []
#     print 'Input = ', xdate
    for i in range(len(xdate)):
        x = re.split('\s+', xdate[i])
#         print 'Split on space = ', x
        date = x[0].split('/')
        timex = x[1].split(':')
        if x[2].lower() == 'pm' and timex[0] != '12':
#             print 'True'
            timex[0] = int(timex[0]) + 12
        else:
            pass
#             print 'False'
#         print 'Date = ', date
#         print 'Time = ', timex
#         print datetime.date(int(date[2]), int(date[0]), int(date[1]))
#         print datetime.time(int(timex[0]), int(timex[1]))
#         print datetime.datetime.combine(datetime.date(int(date[2]), int(date[0]), int(date[1])), datetime.time(int(timex[0]), int(timex[1])))
        finalIndex.append(datetime.datetime.combine(datetime.date(int(date[2]), int(date[0]), int(date[1])), datetime.time(int(timex[0]), int(timex[1]))))
#     finalIndex = finalIndex.tseries.tools.to_datetime()
#     print finalIndex
    return finalIndex
```

```python
def collapse_by_day(df):
    new_df = pd.DataFrame()
    
    new_df['soil_m_mean'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].mean()
#     print new_df
    new_df['soil_m_max'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].max()
    new_df['soil_m_min'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].min()
    new_df['soil_m_std'] = df.groupby(pd.TimeGrouper('1D'))['soil_m'].std()
    return new_df
```

```python
final_df = pd.DataFrame()
df = pd.DataFrame()
for x in range(len(deca_files)):
    for f in os.listdir(deca_files[x]):
        if 'H2M1_2_21Jun14-1804.csv' in f and not f.startswith('.'):
            
            tmp_df = pd.read_csv(join(deca_files[x],f), skiprows = 3, header = None, index_col = 0)
#             tmp_df.set_index(parse_date(tmp_df.index.values))
            df = pd.DataFrame(index = parse_date(tmp_df.index.values))
        
            df1 = df.copy()
            df1['soil_m'] = tmp_df[1].values
            df1 = df1[df1['soil_m'].notnull()]
#             df1.index.freq = '2H'
            print df1.index.freq
            print df1.index
            day_df1 = collapse_by_day(df1)
            day_df1['DateTime'] = day_df1.index.values            
            day_df1['treatment'] = 'Cr'
            day_df1['habitat'] = 'C'
            day_df1['site'], day_df1['soil_type'], day_df1['monitor_station'], day_df1['replicate'] = parse_deca_filename(f)
            day_df1['filename'] = f

            df2 = df.copy()
            df2['soil_m'] = tmp_df[2].values
            df2 = df2[df2['soil_m'].notnull()]
            day_df2 = collapse_by_day(df2)
            day_df2['DateTime'] = day_df2.index.values            
            day_df2['treatment'] = 'C'
            day_df2['habitat'] = 'P'
            day_df2['site'], day_df2['soil_type'], day_df2['monitor_station'], day_df2['replicate'] = parse_deca_filename(f)
            day_df2['filename'] = f
            
            df3 = df.copy()
            df3['soil_m'] = tmp_df[3].values
            df3 = df3[df3['soil_m'].notnull()]
            day_df3 = collapse_by_day(df3)
            day_df3['DateTime'] = day_df3.index.values            
            day_df3['treatment'] = 'H'
            day_df3['habitat'] = 'C'
            day_df3['site'], day_df3['soil_type'], day_df3['monitor_station'], day_df3['replicate'] = parse_deca_filename(f)
            day_df3['filename'] = f
            
            df4 = df.copy()
            df4['soil_m'] = tmp_df[4].values
            df4 = df4[df4['soil_m'].notnull()]
            day_df4 = collapse_by_day(df4)
            day_df4['DateTime'] = day_df4.index.values
            day_df4['treatment'] = 'H'
            day_df4['habitat'] = 'S'
            day_df4['site'], day_df4['soil_type'], day_df4['monitor_station'], day_df4['replicate'] = parse_deca_filename(f)
            day_df4['filename'] = f
    
            final_df = final_df.append(day_df1, ignore_index = True)
            final_df = final_df.append(day_df2, ignore_index = True)
            final_df = final_df.append(day_df3, ignore_index = True)
            final_df = final_df.append(day_df4, ignore_index = True)
final_df.to_csv('by_day_Decagon.csv')
```

```python
def parse_butt(f):
    new_df = pd.read_csv(f, skiprows = 20, header = None)
    return new_df
```

```python
def parse_butt_filename(f):
    f = f.split('_')
    print f
    first = f[0]
    second = f[1]
    site = first[0]
    soil_type = first[1]
    #an I in one of the filenames (soil type), converted to 1
    if soil_type == 'I':
        soil_type = 1
    monitor_station = first[-2:]
    treatment = first[2:4]
    habitat = first[4:]
    return site, soil_type, monitor_station, treatment, habitat
```
